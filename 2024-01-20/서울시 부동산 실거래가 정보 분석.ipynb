# 서울시 부동산 실거래가 정보 분석

```python
import pandas as pd
```

## 데이터 로드

```python
df = pd.read_csv('data.csv')
```

```python
df
```

## 결측치 제거

```python
df.fillna('-')
```

## 건물 용도별로 묶기

```python
df = df.sort_values(by='건물용도')
df.index = [i for i in range(2572)]
```

```python
df
```

## 아파트 관련 데이터만 추출하기

```python
start = 0
end = 0

for i in range(len(df.index)):
    if not start and df['건물용도'][i] == '아파트':
        start = i
    elif start and df['건물용도'][i] != '아파트':
        end = i
        break

apartment = df[start:end]
```

```python
apartment
```

## 자치구별로 묶기

```python
apartment = apartment.sort_values(by='자치구명')
```

```python
apartment
```

## 자치구별 아파트 실거래가 정보 추출

```python
import matplotlib.pyplot as plt
```

```python
apartment.groupby('자치구명').mean()
```

```python
apartment.groupby('자치구명').mean()['물건금액(만원)']
```

### 폰트 깨짐 문제 해결

```python
import matplotlib.font_manager as fm

# 나눔글꼴 경로 설정
# Windows: font_path = 'C:/Windows/Fonts/NanumGothic.ttf'
font_path = '/Library/Fonts/Arial Unicode.ttf'

# 폰트 이름 가져오기
font_name = fm.FontProperties(fname=font_path).get_name()

# 폰트 설정
plt.rc('font', family=font_name)
```

### 그래프 표출

```python
# 그래프 크기 조절
plt.figure(figsize=(20, 5))

# 그래프 x, y값 설정
plt.bar(apartment.groupby('자치구명').mean()['물건금액(만원)'].index, apartment.groupby('자치구명').mean()['물건금액(만원)'])

# 그래프 x축 텍스트 회전
plt.xticks(rotation=45, fontsize=15)

# 그래프 표출
plt.show()
```

## 예산에 맞는 아파트 찾기

```python
budget = int(input('예산 입력 : '))
```

```python
for i in apartment.index:
    if apartment['물건금액(만원)'][i] <= budget:
        print(apartment['건물명'][i:i+1], end='\n\n')
```
