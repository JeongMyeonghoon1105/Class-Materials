# 서울시 부동산 실거래가 정보 분석

```python
import pandas as pd
```

## 데이터 로드

```python
df = pd.read_csv('data.csv')
```

```python
df
```

## 결측치 제거

```python
df.fillna('-')
```

## 건물 용도별로 묶기

```python
df = df.sort_values(by='건물용도')
df.index = [i for i in range(2572)]
```

```python
df
```

## 아파트 관련 데이터만 추출하기

```python
start = 0
end = 0

for i in range(len(df.index)):
    if not start and df['건물용도'][i] == '아파트':
        start = i
    elif start and df['건물용도'][i] != '아파트':
        end = i
        break

apartment = df[start:end]
```

```python
apartment
```

## 자치구별로 묶기

```python
apartment = apartment.sort_values(by='자치구명')
```
