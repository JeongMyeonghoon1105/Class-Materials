## 데이터 로드

## 결측치 제거

```python
df.fillna('-')
```

## 건물 용도별로 데이터프레임 분리

```python
df = df.sort_values(by='건물용도')
```

```python
df
```

```python
start = 0
end = 0

for i in range(len(df.index)):
    if not start and df['건물용도'][i] == '아파트':
        start = i
    if start and df['건물용도'][i] != '아파트':
        end = i
        break

print(start, end)
apartment = df[start:end+1]
```

```python
apartment
```
